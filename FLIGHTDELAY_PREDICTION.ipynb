```python
df = df.drop('Unnamed: 25',axis=1)
df.isnull().sum()
```

```python
df = df[['MONTH','DAY_OF_MONTH','DAY_OF_WEEK','ORIGIN','DEST','CRS_DEP_TIME','ARR_DEL15']]
df.isnull().sum()     # final filtered data set
```

```python
df.head()            
```

```python
df[df.isnull().values.any(axis=1)].head()  #To check if there is any null values
```

```python
df = df.fillna({'ARR_DEL15':1})              
df.iloc[177:185]
```

```python
import math

for index, row in df.iterrows():
    df.loc[index,'CRS_DEP_TIME']=math.floor(row['CRS_DEP_TIME']/100)
    df.head()
```

```python
df = pd.get_dummies(df, columns=['ORIGIN','DEST'])
df.head()
```

```python
# BUILDING A MACHINE LEARNING MODEL
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
train_x, test_x, train_y, test_y = train_test_split(df.drop('ARR_DEL15', axis=1),df['ARR_DEL15'],test_size=0.2, random_state=42)
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
model1 = RandomForestClassifier(random_state=13)
model1.fit(train_x,train_y)
```

```python
predicted = model1.predict(test_x)
model1.score(test_x,test_y)
```

```python
from sklearn.metrics import roc_auc_score
probabilities = model1.predict_proba(test_x)
roc_auc_score(test_y,probabilities[:, 1])
```

```python
# Generating a confusion matrix
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y,predicted)
```

```python
from sklearn.metrics import precision_score
train_predictions = model1.predict(train_x)
precision_score(train_y, train_predictions)
```

```python
# measuring the model's recall score
```

```python
from sklearn.metrics import recall_score
recall_score(train_y, train_predictions)
```
